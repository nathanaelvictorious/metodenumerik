```
import matplotlib.pyplot as plt
from math import sqrt
plt.style.use('seaborn-whitegrid')

def f(x) :
  return (0.5*2.71828**x)-(3/sqrt(x+1))

iteration = []
xArr = []
erArr = []

x1 = 0
x2 = 2
er = 100
i = 0

while er>=0.04 :
  x = x2 - ((((x2)-(x1))*f(x2))/(f(x2)-f(x1)))
  xArr.append(x)

  iteration.append(i)
  i += 1
  
  x1 = x2
  x2 = x
  erArr.append(er)
  er = (abs(x2-x1)/x2)*100
  


plt.plot(iteration, xArr, marker="o")
plt.title("Xr dari Persamaan f(x)=  1/2  e^x-  3/âˆš(x+1)")
plt.xlabel("Pengulangan ke - ")
plt.ylabel("Nilai Xi+1")
plt.show()

plt.plot(iteration, erArr, marker="o", color="red", linestyle="dashed")
plt.xlabel("Pengulangan ke - ")
plt.ylabel("Error (%)")
plt.show()
```
